```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['country'] = df['country'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['director'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['cast'] = df['cast'].fillna('Unknown')
```

```python
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
df.isnull().sum()
```

```python
df['type'].value_counts() #Movies vs TV Shows Analysis
```

```python
df.groupby('release_year')['title'].count() #his shows how many shows were released each
```

```python
df['country'].value_counts().head(10) #Top 10 countries with the most content on Netflix 
```

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce') #Strip the extra spaces first, then convert to datetime.
```

```python
df.isnull().sum()
```

```python
df['year_added'] = df['date_added'].dt.year  #Extract the year for analysis:
```

```python
df.groupby('year_added')['title'].count() #Shows how many shows were added to Netflix each year, which can help identify trends in content acquisition.
```

```python
df['rating'].value_counts() #Most Common Ratings
```

```python
df['listed_in'].value_counts().head(10) #Tfinds the 10 most common genres on Netflix.
```

```python
pd.pivot_table(
    df,
    values='title',
    index='country',
    columns='type',
    aggfunc='count'
) #Creates a pivot table to analyze the distribution of movies and TV shows by country, which can reveal which countries produce more of each type of content.
```

```python
df.sort_values(by='release_year', ascending=False).head(10) #Find latest content.
```

```python
df['content_age'] = df['release_year'].apply(
    lambda x: "Old" if x < 2015 else "New"
) #Classify Old vs New content.
```

```python
df.corr(numeric_only=True) #Correlation Analysis: This can help identify if there's any relationship between the release year, the year added to Netflix, and the content age classification.
```

```python
df.to_csv("clean_netflix_data.csv", index=False)
```
